По данным 7 измерений некоторой величины найдены средняя результатов измерений, равная 30 и выборочная дисперсия, равная 36. Найдите границы, в которых с надежностью 0,99 заключено истинное значение измеряемой величины.

```python
from scipy import stats
import math

# Данные
среднее = 30
выборочная_дисперсия = 36
количество_измерений = 7
доверительная_вероятность = 0.99

# Вычисление стандартной ошибки
стандартная_ошибка = math.sqrt(выборочная_дисперсия / количество_измерений)

# Вычисление интервала
интервал = stats.t.interval(доверительная_вероятность, количество_измерений - 1, loc=среднее, scale=стандартная_ошибка)

print(f"Доверительный интервал: {интервал}")
```

По группе семей с доходом 154 тысяч руб./чел. зафиксированы следующие цифры потребления молока (л) за месяц (на одного человека): 8,3; 8,6; 8,7; 8,8; 9,1; 9,3; 9,4; 13,4; 13,5; 13,8; 13,9; 14,1; 14,3. Найти доверительный интервал для математического ожидания с надежностью 0.95. Выборка произведена из нормальной совокупности.

```python
import numpy as np
import scipy.stats as stats

# Данные
потребление_молока = np.array([8.3, 8.6, 8.7, 8.8, 9.1, 9.3, 9.4, 13.4, 13.5, 13.8, 13.9, 14.1, 14.3])
доверительная_вероятность = 0.95

# Вычисляем среднее и стандартную ошибку
среднее = np.mean(потребление_молока)
стандартная_ошибка = stats.sem(потребление_молока)

# Вычисляем доверительный интервал
интервал = stats.norm.interval(доверительная_вероятность, loc=среднее, scale=стандартная_ошибка)

print(f"Доверительный интервал: {интервал}")
```

С целью размещения рекламы опрошено 420 телезрителей, из которых данную передачу смотрят 170 человек. С доверительной вероятностью γ=0,91 найти долю телезрителей, охваченных рекламой в лучшем случае.

```python
import scipy.stats as stats
import math

n = 420  # размер выборки
p = 170 / 420  # выборочная пропорция

# Находим квантиль нормального распределения для доверительной вероятности 0,91
z = stats.norm.ppf(0.955)

# Вычисляем доверительный интервал
статистическая_ошибка = math.sqrt((p * (1 - p)) / n)
доверительный_интервал = z * статистическая_ошибка

# Вычисляем долю телезрителей, охваченных рекламой
лучший_случай = p + доверительный_интервал
print(f"Доля телезрителей, охваченных рекламой в лучшем случае: {лучший_случай:.4f}")
```

Оценить с надежностью γ = 0.95 математическое ожидание a нормально распределенного признака генеральной совокупности по выборочной средней при помощи доверительного интервала

```python
import numpy as np
import scipy.stats as stats

выборка = np.array([-2, -2, 1, 2, 2, 3, 3, 4, 4, 5])
среднее = np.mean(выборка)
std_error =  np.std(выборка, ddof=1) / np.sqrt(len(выборка))
доверительный_интервал = stats.norm.interval(0.95, loc=среднее, scale=std_error)

print(f"Доверительный интервал для среднего: {доверительный_интервал}")
```
