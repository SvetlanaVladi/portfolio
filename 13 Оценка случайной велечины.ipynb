```python
# Данные из таблицы
experience = [5, 8, 13, 18, 23, 25]   # возьмем средний стаж в каждой группе
workers = [70, 60, 20, 22, 18, 10]

# Вычисление общего количества рабочих и общего стажа
total_workers = sum(workers)
total_experience = sum([experience[i] * workers[i] for i in range(len(experience))])

# Вычисление среднего стажа
avg_experience = total_experience / total_workers
print(f"Средний стаж рабочего: {avg_experience} лет")
```

```python
# Исходные данные
стаж = [5, 7.5, 12.5, 17.5, 22.5, 25]
число_рабочих = [70, 60, 20, 22, 18, 10]

# Уменьшаем частоты в два раза и увеличиваем варианты на две единицы
новые_число_рабочих = [число // 2 for число in число_рабочих]
новый_стаж = [вариант + 2 for вариант in стаж]

# Вычисляем новый средний стаж
новый_средний_стаж = sum([новый_стаж[i] * новые_число_рабочих[i] for i in range(len(новый_стаж))]) / sum(новые_число_рабочих)

print(f"Новый средний стаж рабочего: {новый_средний_стаж} лет")
```

```python
# Время, затраченное на изготовление деталей
время_рабочего_1 = 30  # минут
время_рабочего_2 = 40  # минут

# Вычисляем среднюю затрату времени на изготовление одной детали
среднее_время = (время_рабочего_1 + время_рабочего_2) / 2
print(f"Средняя затрата времени на изготовление одной детали: {среднее_время} минут")
```

```python
# Расстояние и скорость для каждого участка
расстояние_1 = 480  # км
скорость_1 = 60  # км/ч
расстояние_2 = 320  # км
скорость_2 = 80  # км/ч
расстояние_3 = 200  # км
скорость_3 = 50  # км/ч

# Вычисляем среднюю скорость
средняя_скорость = (расстояние_1 + расстояние_2 + расстояние_3) / ((расстояние_1 / скорость_1) 
                    + (расстояние_2 / скорость_2) + (расстояние_3 / скорость_3))
print(f"Средняя скорость рейса: {средняя_скорость} км/ч")
```
