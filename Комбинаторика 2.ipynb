В конференции участвуют 12 французов, 11 россиян, 45 американцев и 32 англичанина. Порядок прочтения докладов определяется жребием. Какова вероятность того, что заключительный доклад будет читаться россиянином?

Чтобы пройти в следующий круг соревнований, футбольной команде нужно набрать хотя бы 4 очка в двух играх. Если команда выигрывает, она получает 3 очка, в случае ничьей — 1 очко, если проигрывает — 0 очков. Найдите вероятность того, что команде удастся выйти в следующий круг соревнований. Считайте, что в каждой игре вероятности выигрыша и проигрыша одинаковы и равны 0,3.

```python
# Равновероятные события. P = m / N (где m - подходящие события, N - всего исходов)
# Всего 8 исходов = 3-3, 3-1, 1-3, 1-1, 0-3, 3-0, 1-0, 0-1
# что бы пройти нужны две игры в сумме чтоб было 4, это три комбинации 3-3, 3-1, 1-3

N = 8
m = 3
p = m / N
print(p)
```

Таня заметила, что в казино “Подкинем” используют неправильную игральную кость (т.е. не у всех граней вероятности выпадения одинаковы). При этом она установила, что вероятность выпадения чётного числа равна 0,6; вероятность выпадения числа, делящегося на 3, равна 0,3; вероятность того, что выпадет 1 или 5, равна 0,22. Какова вероятность того, что на этой игральной кости выпадет число 3?

```python
P_even = 0.6            # Вероятность выпадения чётного числа
P_1_or_5 = 0.22         # Вероятность выпадения числа 1 или 5

P_3 = 1 - P_even - P_1_or_5  # Вероятность выпадения числа 3

print("Вероятность выпадения числа 3 на данной игральной кости:", round(P_3, 2))
```

Компания “Light” изготавливает лампочки. Вероятность того, что готовая лампочка неисправна, равна 0,04. Каждую лампочку дополнительно проверяет упаковщик. Вероятность того, что упаковщик обнаружит (и изымет) неисправную лампочку, равна 0,96. Вероятность того, что упаковщик по ошибке изымет исправную лампочку, равна 0,01. Найдите вероятность того, что случайно выбранная изготовленная лампочка будет изъята упаковщиком.

```python
# Мы можем представить P(B) в виде суммы вероятностей, когда упаковщик изымает лампочку в различных случаях:

P_A = 0.04              # Вероятность того, что лампочка неисправна
P_B_given_A = 0.96      # Вероятность обгаружения и изьятия неисправной лампочки
P_B_given_not_A = 0.01  # Вероятность изъятия лампочки при условии, что лампочка исправная
P_not_A = 1 - P_A       # Вероятность исправной лампочки

P_B = P_B_given_A * P_A + P_B_given_not_A * P_not_A  # Вероятность изъятия лампочки упаковщиком

print("Вероятность изъятия лампочки упаковщиком:", round(P_B, 3))
```
